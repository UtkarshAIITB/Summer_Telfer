```python
def data(i):
  column_names = ['Reserves and funds', 'Total income', 'Paid up equity capital (net of forfeited equity capital)', 'Year', 'Net working capital', 'Total expenses', 'Operating expenses of non-finance cos', 'Total liabilities', 'Total assets', 'Sales', 'Market Capitalisation', 'Net cash inflow or (outflow) from investing activities', 'Net cash flow from operating activities', 'Retained profits/losses during the year', 'Profit after tax', 'Current liabilities & provisions']
  wow = pd.DataFrame(columns = column_names)
  c1 = df.loc[i]
  c1 = c1.to_frame()
  c1 = c1.reset_index(drop = True)
  c1.columns = [c1.loc[0]]
  c1 = c1.drop(0).reset_index(drop = True)
  for m in range(0,10):
    store = pd.DataFrame()
    store = c1.loc[15*m + m%15:15*(m+1) + (m)%15]
    store_transpose = store.T
    store_transpose = store_transpose.reset_index(drop = True)
    store_transpose.columns = column_names
    wow = wow.append(store_transpose)
    wow  = wow.reset_index(drop = True)
  return wow
```

```python
op = pd.DataFrame(columns = ['Reserves and funds', 'Total income', 'Paid up equity capital (net of forfeited equity capital)', 'Year', 'Net working capital', 'Total expenses', 'Operating expenses of non-finance cos', 'Total liabilities', 'Total assets', 'Sales', 'Market Capitalisation', 'Net cash inflow or (outflow) from investing activities', 'Net cash flow from operating activities', 'Retained profits/losses during the year', 'Profit after tax', 'Current liabilities & provisions'])
for i in range(0, df.shape[0]):
  lessgo = data(i)

  for column in lessgo.columns:
    if(column != 'Year'):
      lessgo[column] = lessgo[column].astype('float')

  lessgo.loc[:, 'Company'] = (df.loc[i])[0]
  lessgo.loc[:, 'Equity'] = ( lessgo.loc[:,'Paid up equity capital (net of forfeited equity capital)']) + (lessgo.loc[:, 'Reserves and funds'])

  op = op.append(lessgo)
  op = op.reset_index(drop = True)
```

```python
from sklearn.impute import KNNImputer
```

```python
imp = one_plus.copy()
imp = imp[imp['Company'].str.contains('Bank') == False]
for company in imp['Company'].unique():
  comp = imp[imp['Company'] == company]
  comp = comp.reset_index(drop = True)
  if(  (comp['Sales'].isnull().sum() != comp.shape[0]) and (comp['Market Capitalisation'].isnull().sum() != comp.shape[0]) ):
    pseudo = comp.iloc[:,0:2]
    comp = comp.drop(columns = ['Year', 'Company'])
    knn = KNNImputer(n_neighbors=3,weights='distance')
    comp_trf = knn.fit_transform(comp)
    # print(comp_trf.shape)
    df_imputer = pd.DataFrame(comp_trf, columns = ['Paid up equity capital (net of forfeited equity capital)',
       'Reserves and funds', 'Total income', 'Total expenses',
       'Total liabilities', 'Total assets', 'Current liabilities & provisions',
       'Profit after tax', 'Sales', 'Net cash flow from operating activities',
       'Net cash inflow or (outflow) from investing activities', 'Equity',
       'Retained profits/losses during the year', 'Net working capital',
       'Market Capitalisation', 'Operating expenses of non-finance cos'])
    frames = [pseudo, df_imputer]
    df_all = pd.concat(frames, axis = 1)
    df_khatm = df_khatm.append(df_all)
    df_khatm = df_khatm.reset_index(drop = True) 
```

```python
for i in range(0, df_khatm.shape[0]):
  # EBIT
  df_khatm.loc[i ,'EBIT'] = df_khatm.loc[i, 'Total income'] - df_khatm.loc[i, 'Operating expenses of non-finance cos']
```

```python
# # Altman model ratios

# Working Capital/Total Assets ratio : r1
# Retained Earnings/Total Assets : r2
# Earnings before interest and Tax (EBIT)/ Total Assets Ratio : r3
# Market Value of Equity/Total Liabilities : r4
# Total Sales/Total Assets : r5
```

```python
for i in range(0, df_khatm.shape[0]):
  # r1
  if df_khatm.loc[i, 'Total assets'] != 0 :
    df_khatm.loc[i, 'r1'] = (df_khatm.loc[i, 'Net working capital'])/(df_khatm.loc[i, 'Total assets'])
  else:
    df_khatm.loc[i, 'r1'] = 0

  # r2
  if df_khatm.loc[i, 'Total assets'] != 0 :
    df_khatm.loc[i, 'r2'] = (df_khatm.loc[i, 'Retained profits/losses during the year'])/(df_khatm.loc[i, 'Total assets'])
  else:
    df_khatm.loc[i, 'r2'] = 0
  
  # r3
  if df_khatm.loc[i, 'Total assets'] != 0:
    df_khatm.loc[i, 'r3'] = (df_khatm.loc[i, 'EBIT'])/(df_khatm.loc[i, 'Total assets'])
  else:
    df_khatm.loc[i, 'r3'] = 0
  
  # r4
  if df_khatm.loc[i, 'Total liabilities'] !=0:
    df_khatm.loc[i, 'r4'] = (df_khatm.loc[i, 'Market Capitalisation'])/(df_khatm.loc[i, 'Total liabilities'])
  else:
    df_khatm.loc[i, 'r4'] = 0
  
  # r5
  if df_khatm.loc[i, 'Total assets'] != 0:
    df_khatm.loc[i, 'r5'] = (df_khatm.loc[i, 'Sales'])/(df_khatm.loc[i, 'Total assets'])
  else:
    df_khatm.loc[i, 'r5'] = 0
```

```python
for i in range(0, df_khatm.shape[0]):
  # altman z-score
  df_khatm.loc[i, 'altman z-score'] = 1.2*df_khatm.loc[i, 'r1'] + 1.4*df_khatm.loc[i, 'r2'] + 3.3*df_khatm.loc[i, 'r3'] + 0.6*df_khatm.loc[i, 'r4'] + df_khatm.loc[i, 'r5']
```
