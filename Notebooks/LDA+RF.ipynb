yearwise and yearwise_good are two dataframes consisting of bankrupt and non-bankrupt companies respectively. The values in year_1, year_2 and year_3 are a score which are calculated using LDA (Linear Discriminant Analysis). Lets merge these two dataframes and carry out bankruptcy prediction.

```python
merge = pd.concat( [yearwise, yearwise_good] , ignore_index = True)
print(merge.shape)
merge.head()
```

```python
merge.shape[0] == yearwise.shape[0] + yearwise_good.shape[0]
```

```python
merge['Y'].value_counts()
```

```python
copy = merge.copy()
copy = copy.sample(frac = 1)
copy = copy.reset_index(drop = True)
copy.head()
```

```python
copy['Y'] = copy['Y'].astype('float')
copy.info()
```

```python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from numpy import mean
from sklearn.model_selection import train_test_split
```

```python
y = copy['Y']
X = copy.drop(columns = ['Company', 'Y'])
```

```python
y
```

```python
steps = [('scale', StandardScaler()),('model', RandomForestClassifier())]
pipeline = Pipeline(steps=steps)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(pipeline, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
print('Mean accuracy: %.3f' % mean(scores))
```
